# **Linear Regression (Single Variable) using sklearn**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Step 1: Load dataset
data = pd.read_csv("experience_salary.csv")


X = data[['YearsExperience']]
Y = data['Salary']

data.head()
# print(X)
# print(Y)
```

```python
# Step 2: Create the Linear Regression model

model = LinearRegression()
```

```python
# Step 3: Train the model

model.fit(X,Y)
```

```python
# Step 4: Get the parameters

m= model.coef_[0]
b= model.intercept_
print(f"Trained Model: Salary = {m:.4f} * Experience + {b:.4f}") # y = mx + b
```

```python
# Step 5: Make predictions
Y_pred = model.predict(X)
```

```python

predicted_salary = model.predict([[7]])
print(f"Predicted Salary for 7 years of experience: ${predicted_salary[0]:,.2f}")
```

```python
model.predict([[6]])
```

```python
# Step 6: Plot the regression line with data
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Experience vs. Salary (Linear Regression using sklearn)")
plt.legend()
plt.show()
```
